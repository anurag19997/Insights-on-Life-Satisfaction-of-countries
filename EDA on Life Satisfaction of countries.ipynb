## Correalation between Life Expectency and various factors

```python
for col in df.columns:
    sample_corr = np.corrcoef(df.dropna(subset=[col, 'Life expectancy'])[col], df.dropna(subset=[col, 'Life expectancy'])['Life expectancy'])
    corr = sample_corr[0][1]
    if(corr > 0.5 or corr < -0.5): ## I am taking 0.5 as threshold due to less data
        f = plt.figure()
        ax = f.add_subplot(111) 
        plt.scatter(df[col], df['Life expectancy'])
        plt.text(0.7, 0.2,'corrcoef: '+ str(corr),
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax.transAxes)
        plt.xlabel(col)
        plt.ylabel('Life expectancy')
```
